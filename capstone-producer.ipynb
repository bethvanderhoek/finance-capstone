```
## Simple Producer with JSON message. 

aDict={}
i=0

for i in range(poverty.count()):
    row = poverty.collect()[i]

    aDict = {
        'Id':row[0],
        'State':row[1],
        'Year':row[2],
        'PovertyPop':row[3],
        'PovertyPerc':row[4],

    }
    print(i)
    print(aDict)
    ## Clarification on p.produce
    # produce ( topic, message in JSON)
    p.produce(confluentTopicName,json.dumps(aDict))
    p.flush()
    sleep(5)
```
